```python
# transfer to datetime
PostHistory['CreationDate'] = pd.to_datetime(PostHistory['CreationDate'])

#  closed duplicate
duplicate_closes = PostHistory[
    (PostHistory['PostHistoryTypeId'] == 10) &
    (PostHistory['Text'].str.contains("Duplicate", case=False, na=False))
]

# 
duplicate_closes['Year'] = duplicate_closes['CreationDate'].dt.year
duplicate_closes.groupby('Year').size().plot(kind='bar', figsize=(12,5))
plt.ylabel("Number of duplicates closed")
plt.title("Duplicate question closures per year")
plt.show()
```
